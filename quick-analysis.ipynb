```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('silver.csv')

df[:5]
```

### Covert date

```python
df['date'] = pd.to_datetime(df['date'])
```

### Magnitude analysis

```python
df['mag_rounded_down'] = df['mag'].astype(int)

df.groupby('mag_rounded_down').agg({'id': 'count'})
```

### Calculate energy

```python
df['energy'] = 5.24
df['energy'] += 1.44 * df['mag']
df['energy'] = np.power(10, df['energy'])
```

### Count through years

```python
plt.hist(df['date'], bins = 500);
```
